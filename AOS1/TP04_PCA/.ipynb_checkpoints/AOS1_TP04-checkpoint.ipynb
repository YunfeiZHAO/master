## PCA for dimension reduction

```python
n, p = 100, 15
```

```python
X = np.random.normal(size=(n, p))
```

```python
X.shape
```

```python
X_mean = np.mean(X, axis=0)
X_mean.shape
```

```python
X_mean
```

```python
# broadcasting, mean 100 times
X0 = X - X_mean
np.mean(X0)
```

```python
# unbiased variance-covariance estimator
V = (1/(n - 1)) * X0.T @ X0
V.shape
```

```python
vp, U = linalg.eigh(V)
```

```python
# help(linalg.eig)
```

```python
vp
```

```python
U.shape
```

```python
X0 @ U
```

```python
np.linalg.norm(U[:, 0])
```

```python
Xpca = X0 @ U
```

```python
n/(n-1) * Xpca.std(axis=0) ** 2
```

```python
vp
```

```python
# don't need to 
pca = PCA()
pca.fit(X)
```

```python
pca.explained_variance_
```

```python
pca.components_.shape
```

```python
U.shape
```

```python
pca.components_
```

```python
U
```

## 2, PCA for dimension reduction

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
pca = PCA()
```

```python
pca.fit(X)
plt.bar(range(1, X.shape[1] + 1), pca.explained_variance_)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
cum = np.cumsum(pca.explained_variance_ratio_)
plt.bar(range(1, X.shape[1] + 1), cum)
```

## We take the fisrt two principle components by exprience

## do the cross validation for the number of components

```python
pca = PCA()
lin = LinearRegression()
pca_lin = Pipeline([("pca", pca), ("lin", lin)])

clf = GridSearchCV(
    estimator=pca_lin,
    scoring="neg_mean_squared_error",
    cv=10,
    iid=False,
    param_grid=dict(pca__n_components=range(1, X.shape[1] + 1)),
)
clf.fit(X, y)
```

```python
clf.best_estimator_
```

```python
clf.best_params_
```

```python
X.std(axis=0)
```

```python
X0 = X - X.mean(axis=0)
X_scaled = X0 / X0.std(axis=0)
```

```python
X_scaled.std(axis=0)
```

```python
X_scaled.mean(axis=0)
```

```python
clf.fit(X_scaled, y)
```

```python
clf.best_params_
```

```python
pca.fit(X)
plt.bar(range(1, X.shape[1] + 1), pca.explained_variance_)
```

```python
pca.fit(X_scaled)
plt.bar(range(1, X_scaled.shape[1] + 1), pca.explained_variance_)
```
